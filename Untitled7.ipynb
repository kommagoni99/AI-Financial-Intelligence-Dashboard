```python
# ===============================
# AI FINANCIAL INTELLIGENCE SYSTEM
# ===============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score
import random

np.random.seed(42)

# -------------------------------
# 1. GENERATE SYNTHETIC DATA
# -------------------------------

def generate_financial_data(n=1000):
    income = np.random.normal(50000, 15000, n)
    savings = income * np.random.uniform(0.1, 0.5, n)
    expenses = income - savings
    age = np.random.randint(22, 60, n)
    investments = savings * np.random.uniform(0.3, 0.9, n)
    risk_score = np.random.randint(1, 10, n)

    data = pd.DataFrame({
        "Income": income,
        "Savings": savings,
        "Expenses": expenses,
        "Age": age,
        "Investments": investments,
        "RiskScore": risk_score
    })
    return data

df = generate_financial_data()

print("Dataset Preview:")
print(df.head())

# -------------------------------
# 2. RISK CLASSIFICATION MODEL
# -------------------------------

X = df[["Income", "Savings", "Investments", "Age"]]
y = (df["RiskScore"] > 5).astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

predictions = rf.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print("\nRisk Classification Accuracy:", accuracy)

# -------------------------------
# 3. INVESTMENT PREDICTION MODEL
# -------------------------------

X_reg = df[["Income", "Savings", "Age"]]
y_reg = df["Investments"]

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.2)

model = LinearRegression()
model.fit(X_train_r, y_train_r)

y_pred = model.predict(X_test_r)
mse = mean_squared_error(y_test_r, y_pred)

print("Investment Prediction MSE:", mse)

# -------------------------------
# 4. VISUAL ANALYSIS
# -------------------------------

plt.figure(figsize=(8,6))
sns.scatterplot(x="Income", y="Investments", data=df)
plt.title("Income vs Investment Distribution")
plt.show()

fig = px.histogram(df, x="RiskScore", title="Risk Score Distribution")
fig.show()

# -------------------------------
# 5. PORTFOLIO SIMULATION
# -------------------------------

def simulate_portfolio(initial_amount=10000, years=5):
    growth_rates = np.random.normal(0.08, 0.15, years)
    values = [initial_amount]

    for rate in growth_rates:
        values.append(values[-1] * (1 + rate))

    return values

portfolio = simulate_portfolio()

plt.figure(figsize=(8,6))
plt.plot(portfolio)
plt.title("Simulated Portfolio Growth")
plt.xlabel("Years")
plt.ylabel("Portfolio Value")
plt.show()

# -------------------------------
# 6. USER INTERACTIVE SECTION
# -------------------------------

print("\n--- Financial Risk Analyzer ---")

income = float(input("Enter your income: "))
savings = float(input("Enter your savings: "))
age = int(input("Enter your age: "))
investments = float(input("Enter your investment amount: "))

user_data = np.array([[income, savings, investments, age]])
risk_prediction = rf.predict(user_data)

if risk_prediction[0] == 1:
    print("⚠ High Risk Profile")
else:
    print("✅ Low Risk Profile")

print("Predicted Ideal Investment:",
      model.predict([[income, savings, age]])[0])
```
